# Task Teaching Queue (list-blacked)

```python
# Teaching Queue (List-backed) with step-by-step outputs

class Queue:
    def __init__(self):
        self.items = []  # internal list to hold elements
        print(f"Init ->", self.items)

    def is_empty(self):
        return len(self.items) == 0  # empty if length == 0

    def enqueue(self, x):
        self.items.append(x)  # append at end (rear)
        print(f"enqueue({x}) ->", self.items)

    def dequeue(self):
        if self.is_empty():
            print("dequeue() -> Underflow (None) |", self.items) # signal underflow
            return None
        # remove & return element at index 0 (front)
        val = self.items.pop(0)  
        print(f"dequeue() -> {val} |", self.items)
        return val

    def front(self):
        if self.is_empty():
            print("front() -> None |", self.items)
            return None
        # peek front without removing
        print(f"front() ->", self.items[0], "|", self.items) 
        return self.items[0]

# quick test with step-by-step tracing
q = Queue()
q.enqueue(10)
q.enqueue(20)
# Use the underscore for a result we don't care about, common in Python
_ = q.dequeue()  # expect 10
_ = q.front()    # expect 20
print('is_empty ->', q.is_empty())  # expect False
```
