# Hospital patient managment sysytem

```python
import heapq
from collections import deque

class Patient:
    def __init__(self, name, priority):
        self.name = name
        self.priority = priority

    def __repr__(self):
        return f"Patient(name={self.name}, priority={self.priority})"

class HospitalQueue:
    def __init__(self, history_limit=10):
        self.priority_queue = []
        self.history = deque(maxlen=history_limit)
        self.counter = 0

    def add_patient(self, name, priority):
        heapq.heappush(self.priority_queue, (-priority, self.counter, Patient(name, priority)))
        self.counter += 1

    def serve_patient(self):
        if not self.priority_queue:
            print("No patients to serve.")
            return None
        neg_priority, _, patient = heapq.heappop(self.priority_queue)
        self.history.append(patient)
        return patient

    def search_patient(self, name):
        for item in self.priority_queue:
            if item[2].name == name:
                return item[2]
        return None

    def get_history(self):
        return list(self.history)

def main():
    hq = HospitalQueue(history_limit=5)
    hq.add_patient("Rehan", 3)
    hq.add_patient("Ali", 1)
    hq.add_patient("Faiz", 5)
    hq.add_patient("Rehman", 2)
    hq.add_patient("Usman", 4)

    print("Serving patients:")
    print(hq.serve_patient())
    print(hq.serve_patient())
    print(hq.serve_patient())

    print("\nSearch for 'Ali':", hq.search_patient("Ali"))

    print("\nHistory of served patients:")
    for patient in hq.get_history():
        print(patient)

if __name__ == "__main__":
    main()
```
